```python
# Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")

# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)

# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

## D1: Determine the Summary Statistics for June

```python
# 1. Import the sqlalchemy extract function.
from sqlalchemy import extract

# 2. Write a query that filters the Measurement table to retrieve the temperatures for the month of June. 
June_temp = session.query(Measurement.date, Measurement.tobs).filter(extract('month', Measurement.date) == 6)
```

```python
#  3. Convert the June temperatures to a list.
June_temp = session.query(Measurement.date, Measurement.tobs).filter(extract('month', Measurement.date) == 6).all()
```

```python
print(June_temp)
```

```python
# 4. Create a DataFrame from the list of temperatures for the month of June. 
df_june_temp = pd.DataFrame(June_temp, columns=['date','June Temps'])
```

```python
print (df_june_temp)
```

```python
# 5. Calculate and print out the summary statistics for the June temperature DataFrame.
df_june_temp.describe()
```

## D2: Determine the Summary Statistics for December

```python
# 6. Write a query that filters the Measurement table to retrieve the temperatures for the month of December.
Dec_temp = session.query(Measurement.date, Measurement.tobs).filter(extract('month', Measurement.date) == 12)
```

D3: Statistical Analysis

```python
#Query to retrieve the data and precipitation scores for the month of June
June_temp_prcp = session.query(Measurement.date, Measurement.prcp).filter(extract('month', Measurement.date) == 6).all()
df_june_temp_prcp = pd.DataFrame(June_temp_prcp, columns=['date','PRCP in June'])
df_june_temp_prcp.set_index(df_june_temp_prcp['date'], inplace=True)
```

```python
df_june_temp_prcp.describe()
```

```python
df_june_temp.plot.hist(bins=22)
plt.tight_layout()
```

```python
#Query to retrieve the data and precipitation scores for the month of December
Dec_temp_prcp = session.query(Measurement.date, Measurement.prcp).filter(extract('month', Measurement.date) == 12).all()
df_dec_temp_prcp = pd.DataFrame(Dec_temp_prcp, columns=['date','PRCP in December'])
df_dec_temp_prcp.set_index(df_dec_temp_prcp['date'], inplace=True)
```

```python
df_dec_temp_prcp.describe()
```

```python
df_dec_temp.plot.hist(bins=22)
plt.tight_layout()
```
